```python
data['Popularity'] = pd.to_numeric(data['Popularity'])
data['Popularity'].dtypes
```

The vote_average column catagorized into popularity

```python
edge =[data['Vote_Average'].describe()['min'],
       data['Vote_Average'].describe()['25%'],
        data['Vote_Average'].describe()['50%'],
        data['Vote_Average'].describe()['75%'],
        data['Vote_Average'].describe()['max']]
labels=['not_popular','below_avg','average','popular']
data['Vote_Average'] = pd.cut(data['Vote_Average'] , bins =edge , labels = labels, duplicates = 'drop')
data['Vote_Average']
```

#We would like to split the genres into a list then explore our dataframe to have on;y one genre per row for each movie
